```python
import re
import os
```

```python
path_openclaw_source = '/home/emily/Desktop/packages/openclaw'
filename_append_to_Dockerfile = '/home/emily/Desktop/projects/operation-clusterfuck/ai/agentic_ai/openclaw/config/append-to-Dockerfile'


list_wrapper_commands = [
    '#!/bin/bash',
    'source $PYTHON_VIRTUAL_ENV/bin/activate',
    'python3 /home/node/.openclaw/workspace/test.py',
]

    
```

```python
os.system('cp ' + path_openclaw_source + '/Dockerfile ' + path_openclaw_source + '/Dockerfile.bak')
```

```python
with open(path_openclaw_source + '/Dockerfile') as f:
    list_original = [x.rstrip() for x in f.readlines()]
```

```python
last_cmd = None
last_cmd_i = None
for i, line in enumerate(list_original):
    if line.strip().find('CMD ') == 0:
        last_cmd = line.strip()
        last_cmd_i = i

last_cmd = re.sub(r'\s+', ' ', last_cmd).strip()

print(last_cmd)
```

```python
new_cmd = None
if last_cmd.find('CMD [') == 0:
    cmd_part_list = eval(last_cmd.replace('CMD ', ''))
    new_cmd = ' '.join(cmd_part_list)
else:
    new_cmd = last_cmd.replace('CMD ', '')

new_cmd
```

```python
list_final_result = []
```

```python
for i, line in enumerate(list_original):
    if i != last_cmd_i:
        list_final_result.append(line)
```

```python
for i in range(0, 5):
    list_final_result.append('')
```

```python
with open(filename_append_to_Dockerfile) as f:
    for line in f.readlines():
        list_final_result.append(line.rstrip())
```

```python
for i in range(0, 5):
    list_final_result.append('')
```

```python
list_wrapper_commands.append(new_cmd)
```

```python
with open(path_openclaw_source + '/python_enablement_wrapper.sh', 'w') as f:
    f.write('\n'.join(list_wrapper_commands) + '\n')
```

```python
list_final_result.append('COPY python_enablement_wrapper.sh /home/node/python_enablement_wrapper.sh')
list_final_result.append('RUN chmod +x /home/node/python_enablement_wrapper.sh')
list_final_result.append('CMD ["/home/node/python_enablement_wrapper.sh"]')
```
